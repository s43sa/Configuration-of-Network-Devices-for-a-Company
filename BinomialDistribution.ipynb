```python
#Import the necessary libraries
import numpy as np #library used for working with arrays
import pandas as pd #library used for data manipulation and analysis
import matplotlib.pyplot as plt #library used for plot and visualizations
import seaborn as sms #library used for visualizations
import scipy.stats as stats #this library contains a large number of probability distributions
%matplotlib inline
```

```python
#Define the parameters of the Binomial Distribution
n = 10  # Number of trials
p = 0.8  # Probability of success in each trial
```

```python
#Generate a random sample
k= np.arange(0,11) #0 to 10
k
```

```python
#Import the required function
from scipy.stats import binom

#Generate the probability distribution
binomial = binom.pmf(k=k,n=n,p=p)
```

```python
#Plot the distribution
plt.bar(k,binomial) #make a bar plot
plt.xlabel('Number of Successes') #set the x-axis label
plt.ylabel('Probability of Successes') #set the y-axis label
plt.title(f'Binomial Distribution (n={n}, p={p})') #set the title
plt.show() #display the plot
```
